```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes (Matrix characters)
characters = ["Neo", "Morpheus", "Trinity", "Agent Smith"]
G.add_nodes_from(characters)

# Add edges with corrected relationship labels
relationships = {
    ("Morpheus", "Neo"): "mentors",
    ("Neo", "Trinity"): "loves",
    ("Neo", "Agent Smith"): "fights",
    ("Morpheus", "Trinity"): "trusts",
    ("Agent Smith", "Neo"): "hates"
}

# Add edges to the graph
G.add_edges_from(relationships.keys())

# Position the nodes for visualization
pos = nx.spring_layout(G)

# Draw the nodes and edges
plt.figure(figsize=(10, 8))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color='lightblue',
    node_size=2000,
    font_size=12,
    font_color='darkblue',
    font_weight='bold',
    arrowsize=15,
    edge_color='gray'
)

# Add edge labels
nx.draw_networkx_edge_labels(
    G,
    pos,
    edge_labels=relationships,
    font_size=10,
    font_color='red'
)

plt.title("Directed Relationships Between Matrix Characters", fontsize=14)
plt.show()
```
